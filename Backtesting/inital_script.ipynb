```python
ticker_symbol = "SPY"

SPY_data = yf.download(
    ticker_symbol,
    start="1993-01-01",
    end="2026-02-17",
    auto_adjust=False,
    multi_level_index=False 
)

SPY_data = SPY_data[["Close"]].reset_index()
SPY_data.to_csv("SPY_prices.csv", index=False)
SPY_data.head(10)
```
