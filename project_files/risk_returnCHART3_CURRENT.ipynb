This is the data processing file to manage the date for this chart.

 The chart is to show the risk/returns annualised for bitcoin and gold for the past 9 years. 

```python
import pandas as pd
import requests
```

```python
#Loading in the daily gold price data
df = pd.read_csv("https://raw.githubusercontent.com/HerbDavis20/HerbDavis20.github.io/refs/heads/main/project_files/gold_prices.csv")

df.head(10)
```

```python
#renaming the columns to make it more clear 
df = df.rename(columns={
    'USD/Gold per ounce': 'date',
    'price': 'Price(USD$ / Ounce)'
})
df.head(10)



```

```python

#removing uneccesary columns
df = df[['date', 'Price(USD$ / Ounce)']]
df.head(10)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')
df.head(10)
```

Now that the gold price data is in a cleaner and more managable form I can remove some of the data as I only need price data for the past 9 years for this chart

```python
# Filter rows where the 'date' is greater than or equal to 2014-01-01
df = df[df['date'] >= '2015-01-01']

# Display the first 10 rows to verify
df.head(10)
```

I also want to reset the index back to zero to prevent any issues with merging down the line. 

```python
df.reset_index(drop=True, inplace=True)
df.head(10)
```

Now that I have the appropriate gold data in, I need to load in the bitcoin data and prepeare it in the same way I did for the gold data. 

```python
#this is the API link for the Bitcoin data that I am using.
url = "https://eco-cors-proxy.netlify.app/proxy?url=https%3A%2F%2Fapi.stlouisfed.org%2Ffred%2Fseries%2Fobservations%3Fseries_id%3DCBBTCUSD%26api_key%3D22ee7a76e736e32f54f5df0a7171538d%26file_type%3Djson"

#this get the data out of the API url 
response = requests.get(url)
data = response.json()

#In the api, the data I want is under the heading of "observations" so i want to extract that and convert it to a dataframe 
df_observations = pd.DataFrame(data['observations'])

# Keep only the 'date' and 'value' columns
dfBTC = df_observations[['date', 'value']]
dfBTC.head(10)
```

```python
dfBTC= dfBTC[dfBTC['value'] != "."]

dfBTC.head(10) #displaying some of the dataframe to make sure that removing the null data has worked.
```

```python
# Filter rows where the 'date' is greater than or equal to 2014-01-01
dfBTC = dfBTC[dfBTC['date'] >= '2015-01-01']

# Display the first 10 rows to verify
dfBTC.head(30)
```

I also want to alter the column name to make it clearer what it is and I will do the same for the gold dataframe 

```python
#renaming for BTC
dfBTC = dfBTC.rename(columns={
    'value': 'BTC_USD$'
})
dfBTC.head(10)


#renaming for Gold 
df = df.rename(columns={
    'value': 'Gold USD/ounce'
})
df.head(10)
```

```python
#standardising date format to make it easier to process later on 
dfBTC['date'] = pd.to_datetime(dfBTC['date'], format='%Y-%m-%d', errors='coerce')
dfBTC.head(10)
```

```python
#also want to do the same again to the gold data to ensure that the dates are in the exact same format
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d', errors='coerce')
df.head(10)
```

```python
#resetting the index again to make sure that does not cause issues later on
dfBTC.reset_index(drop=True, inplace=True)
dfBTC.head(10)
```

Now I want to being grouping the data by year and starting to calculate the annual returns and volatility which is done through calculating the standard deviation. I will start with doing **gold**. 

```python
#this creates another column that extracts just the year from the date.
df['year'] = df['date'].dt.year

#this creates another column that calculates the daily return from the prices
df['daily_return'] = df['Price(USD$ / Ounce)'].pct_change()

gold_df = df.dropna(subset=['daily_return'])


annual_returnsG = df.groupby('year').agg(
    avg_daily_return=('daily_return', 'mean'),
    num_days=('daily_return', 'count')  # Count of valid daily returns
).reset_index()

# Calculates return for each year. this is one of the variables on my chart. 
annual_returnsG['annual_return'] = (1 + annual_returnsG['avg_daily_return']) ** annual_returnsG['num_days'] - 1

annual_returnsG.head(10)




```

Now that I have the returns calculated I need to calculate the Risk which I will do by calculating the standard deviation of daily return which I have already cacluated, and multiplying it by SQRT(the number of valid days in the year)

```python
dev_returnsG = df.groupby('year').agg(
    avg_daily_return=('daily_return', 'mean'),
    num_days=('daily_return', 'count'),
    daily_std_dev=('daily_return', 'std'),  # Daily standard deviation
).reset_index()


# risk = daily std dev * sqrt(number of days)
dev_returnsG['risk'] = dev_returnsG['daily_std_dev'] * (dev_returnsG['num_days'] ** 0.5)

dev_returnsG
```

I am going to use this dataframe as the final one that I will export, so I want to merge in the annual return data back in as this is to be used on one of the axis of my chart. 

```python
#merges the annual_returns column back into this dataframe from the annual_returnsG df 
dev_returnsG = dev_returnsG.merge(annual_returnsG[['year', 'annual_return']].drop_duplicates(), on='year', how='left')
dev_returnsG
```

Now that I have the data I need in this dataframe for gold, I can now remove the columns that wont be needed when making the chart. 

```python
dev_returnsG = dev_returnsG[['year', 'annual_return', 'risk']]
dev_returnsG
```

this is the final data that i need for gold so can now export it as a csv. 

```python
dev_returnsG.to_csv('Gold_data_CHART3.csv', index=False)

print(dev_returnsG.head())
dev_returnsG
```
