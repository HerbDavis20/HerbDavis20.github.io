```python

df = pd.read_csv('CorrolationBTC_GOLD_BTC_VIX_dataFINAL(7).csv')
df
```

```python
df1 = df[['date', 'Rolling_Corr_GOLD_BTC', 'Rolling_Corr_VIX_BTC']]


df1
```

```python
import pandas as pd

df1['date'] = pd.to_datetime(df1['date'])

df1['year'] = df1['date'].dt.year

corrAvg = df1.groupby('year')[['Rolling_Corr_GOLD_BTC', 'Rolling_Corr_VIX_BTC']].mean()

corrAvg = corrAvg[corrAvg.index != 2014]
corrAvg
```

```python
corrAvg = corrAvg.drop(columns=['index'])

# Rename the remaining columns
corrAvg.columns = ['Year', 'Rolling_Corr_GOLD_BTC', 'Rolling_Corr_VIX_BTC']

# Verify the structure
print(corrAvg.head())
```

```python
# Reset index so 'year' is a regular column
corrAvg.reset_index(drop=True, inplace=True)
corrAvg.head(10)

# Rename the columns to ensure proper headers
corrAvg.columns = ['Year', 'Rolling_Corr_GOLD_BTC', 'Rolling_Corr_VIX_BTC']
corrAvg
```
