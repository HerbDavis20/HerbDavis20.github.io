```python
import yfinance as yf
```

First I will import the data on VIX index 

```python
#this defines the asset you want to look at then fetches the data and displays it in a dataframe 
ticker_symbol = '^VIX'

vix_data = yf.download(ticker_symbol, start='2016-01-01', end='2024-12-03')

vix_data.head(10)
```

When I began trying to filter some of the data, I was running into a lot of issued that stemmed from the formatting of the heading columns. so this code sorts the headings out and gets rid of the double stack headings. 

```python
print(vix_data.columns)
```

```python
vix_data.columns = vix_data.columns.get_level_values(0)
vix_data.reset_index(inplace=True)
```

```python

vix_data.columns = ['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']

vix_data = vix_data[['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']]
vix_data.head(10)
```

I also don't need all of this data, the only data I need is the data column and the Close column. 

```python
#this code keeps only those columns. 
vix_close = vix_data[['Date', 'Close']]
vix_close.head(10)
```

When trying to filter this data I was running into issues of NaN. The following code changes the format of the data I am using into to float form, so they are now defined as a number instead of a string, meaning it can be more easily filtered and altered. 

```python
vix_close['Close'] = vix_close['Close'].astype(float)
vix_close.head(10)
```

Now I need to begin filtering it to the data that I require for this chart. I want to find all the datapoints that are >25. I only need the values where it first hits 25 however, so any days where it is consecutively over 25 I want to remove. 

```python

# Filter to keep only the first instance where Close >= 30 and skip consecutive occurrences
filtered_vix_close = vix_close[(vix_close['Close'] >= 25) & 
                               (vix_close['Close'].shift(fill_value=0) < 25)]

filtered_vix_close.head(30)
```

Now we have the data we need for the VIX indicator I can load in the data for Bitcoin. 

```python
#this defines the asset you want to look at then fetches the data and displays it in a dataframe 
ticker2 = 'BTC-USD'

btc_data = yf.download(ticker2, start='2016-01-01', end='2024-12-03')

btc_data.head(10)
```

first I want to alter the column names as this caused lots of issues when managing the VIX data earlier. 

```python
# Flatten the headings so that its not across multiple levels
btc_data.columns = btc_data.columns.get_level_values(0)
print(btc_data.columns)

btc_data.reset_index(inplace=True)
```

```python

btc_data.columns = ['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']

btc_data = btc_data[['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']]

btc_data.head(10)

```

I want to remove the unccessary columns. 

```python
btc_close = btc_data[['Date', 'Close']]
btc_close.head(10)
```

```python
filtered_vix_close.head(10)
```

Now I want to begin creating the merged dataframes to calcualte the percentage price changes for bitcoin in 1-day, 7-day, and 30-day periods

```python
#this merges the vix data, for all the values where it breaks 25 with the corresponding bitcoin price for that day where it does break 25. 

merged_1 = pd.merge(btc_close, filtered_vix_close, on='Date', how='inner')

merged_1 = merged_1.rename(columns={
    'Close_x': 'BTC_Close',
    'Close_y': 'VIX_Close'
})

merged_1.head(20)
```

```python
# Add a new column for BTC_Close value one day before the 'Date'
btc_close['Next_Day'] = btc_close['Date'] + pd.Timedelta(days=-1)

# Rename the 'Close' column for clarity
btc_close_renamed = btc_close.rename(columns={'Close': 'Next_Day_BTC_Close'})

# Create a new DataFrame with the merged data
merged_1day = pd.merge(
    merged_1,
    btc_close_renamed[['Next_Day', 'Next_Day_BTC_Close']],
    left_on='Date',
    right_on='Next_Day',
    how='left'
)

# Drop the auxiliary 'Next_Day' column from the new DataFrame to keep it clean
merged_1day = merged_1day.drop(columns=['Next_Day'])

# Display the first 20 rows of the new DataFrame
merged_1day.head(20)
```

Now I want to calculate the percentage change from the BTC_Close value to the Next_Day_BTC_Close as this will go into one of the buckets for the histogram. 

```python
merged_1day['Change_BTC'] = (
    (merged_1day['Next_Day_BTC_Close'] - merged_1day['BTC_Close']) / merged_1day['BTC_Close']
) * 100

merged_1day.head(10)
```

Now I am going to extract just the change_BTC column and data into a new dataframe as this is all I need for my chart and it will make it easier to manage. 

```python
DAYS1 = merged_1day[['Date', 'Change_BTC']]

DAYS1.head(20)


```

Now I want to repeat the same process for 7 days and 30 days, fetching the bitcoin price 7 and 30 days after the breaking of 25 on the VIX, and working out the percentage change in price from that day to 7 and 30 days after. 

```python
merged_1.head(10)
```

```python
# Add a new column for BTC_Close value x (1,7 or 30)  days after the btc_close date
btc_close['Next_7Day'] = btc_close['Date'] + pd.Timedelta(days=-7)

# Rename to make it clearer
btc_close_renamed = btc_close.rename(columns={'Close': 'Next_7Day_BTC_Close'})

# Creates a new df with the merged data
merged_7day = pd.merge(
    merged_1,
    btc_close_renamed[['Next_7Day', 'Next_7Day_BTC_Close']],
    left_on='Date',
    right_on='Next_7Day',
    how='left'
)

# gets rid of the Next_Day column from the new df to keep it clean as this was just used for the calculation
merged_7day = merged_7day.drop(columns=['Next_7Day'])
merged_7day.head(20)
```

```python
#calculates the percentage change and creates a new column for the percentage change.
merged_7day['Change_BTC'] = (
    (merged_7day['Next_7Day_BTC_Close'] - merged_7day['BTC_Close']) / merged_7day['BTC_Close']
) * 100

merged_7day.head(10)
```

```python
#takes the percentage change and date into a seperate datafrme just to make it clear where the finished processed data is. 
DAYS7 = merged_7day[['Date', 'Change_BTC']]

DAYS7.head(20)


```

Now that 1 day and 7 day have been done I need to take the same steps to get the percentage change for 30 days after the breaking of 25 on the vix index. 

```python
# Add a new column for BTC_Close value x (1,7 or 30)  days after the btc_close date
btc_close['Next_30Day'] = btc_close['Date'] + pd.Timedelta(days=-30)

# Rename to make it clearer
btc_close_renamed = btc_close.rename(columns={'Close': 'Next_30Day_BTC_Close'})

# Creates a new df with the merged data
merged_30day = pd.merge(
    merged_1,
    btc_close_renamed[['Next_30Day', 'Next_30Day_BTC_Close']],
    left_on='Date',
    right_on='Next_30Day',
    how='left'
)

# gets rid of the Next_Day column from the new df to keep it clean as this was just used for the calculation
merged_30day = merged_30day.drop(columns=['Next_30Day'])
merged_30day.head(20)
```

```python
#calculates the percentage change and creates a new column for the percentage change.
merged_30day['Change_BTC'] = (
    (merged_30day['Next_30Day_BTC_Close'] - merged_30day['BTC_Close']) / merged_30day['BTC_Close']
) * 100

merged_30day.head(10)
```

```python
#takes the percentage change and date into a seperate datafrme just to make it clear where the finished processed data is. 
DAYS30 = merged_30day[['Date', 'Change_BTC']]

DAYS30.head(20)


```

Now I have all the data I need in the dataframes DAYS1, DAYS7, AND DAYS30.

To make the chart plotting easier I will do the sorting into buckets now. To do so I want to sort the data in to groups as follows:
- <-10%
- -10% to -5%
- -5% to 0%
- 0% to 5%
- 5% to 10%
- (>10%)

```python
# Define the bins and the names of the bins
bins = [-float('inf'), -10, -5, 0, 5, 10, float('inf')]
labels = ['<-10%', '-10% to -5%', '-5% to 0%', '0% to 5%', '5% to 10%', '>10%']

# sorts each value in DAYS1 into its appropriate bucket in a new column called Change_BTC_Bucket
DAYS1['Change_BTC_Bucket'] = pd.cut(DAYS1['Change_BTC'], bins=bins, labels=labels)

#counts the number of data points in each bucket 
sortcount = DAYS1['Change_BTC_Bucket'].value_counts().sort_index()

# creates a new dataframe called DAYS1_FX which contains just the buckets and the frequency for each bucket. this is the final data that is needed for my chart. 
DAYS1_FX= pd.DataFrame({
    'Bucket': sortcount.index,
    'Frequency': sortcount.values
})
DAYS1_FX
```

Now that I have done this for the DAYS1 dataframe I can repeat this same step for DAYS7 and DAYS30 dataframes before exporting them all. 

```python
#this is just a repeat of the code before to get the frequency of all the bins, this time for the DAYS7 dataframe. 

bins = [-float('inf'), -10, -5, 0, 5, 10, float('inf')]
labels = ['<-10%', '-10% to -5%', '-5% to 0%', '0% to 5%', '5% to 10%', '>10%']

DAYS7['Change_BTC_Bucket'] = pd.cut(DAYS7['Change_BTC'], bins=bins, labels=labels)

sortcount = DAYS7['Change_BTC_Bucket'].value_counts().sort_index()

DAYS7_FX= pd.DataFrame({
    'Bucket': sortcount.index,
    'Frequency': sortcount.values
})
DAYS7_FX
```

```python
#this is just a repeat of the code before to get the frequency of all the bins, this time for the DAYS30 dataframe. 

bins = [-float('inf'), -10, -5, 0, 5, 10, float('inf')]
labels = ['<-10%', '-10% to -5%', '-5% to 0%', '0% to 5%', '5% to 10%', '>10%']

DAYS30['Change_BTC_Bucket'] = pd.cut(DAYS30['Change_BTC'], bins=bins, labels=labels)

sortcount = DAYS30['Change_BTC_Bucket'].value_counts().sort_index()

DAYS30_FX= pd.DataFrame({
    'Bucket': sortcount.index,
    'Frequency': sortcount.values
})
DAYS30_FX
```

Now I have the final data for all 3 charts in the concat, I can export the dataframes as csv files. 

```python
DAYS1_FX.to_csv('DAYS1_CHART4.csv', index=False)
DAYS7_FX.to_csv('DAYS7_CHART4.csv', index=False)
DAYS30_FX.to_csv('DAYS30_CHART4.csv', index=False)


```
