This is the data processing file to manage the date for this chart.

 The chart is to show the risk/returns annualised for bitcoin and gold for the past 9 years. 

```python
import pandas as pd
import requests
```

```python
#Loading in the daily gold price data
df = pd.read_csv("https://raw.githubusercontent.com/HerbDavis20/HerbDavis20.github.io/refs/heads/main/project_files/gold_prices.csv")

df.head(10)
```

```python
#renaming the columns to make it more clear 
df = df.rename(columns={
    'USD/Gold per ounce': 'date',
    'price': 'Price(USD$ / Ounce)'
})
df.head(10)


#removing uneccesary columns
df = df[['date', 'Price(USD$ / Ounce)']]
df.head(10)

#standardising date format to make it easier to process later on 
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')

df.head(10)
```

Now that the gold price data is in a cleaner and more managable form I can remove some of the data as I only need price data for the past 10 years for this chart

```python
# Filter rows where the 'date' is greater than or equal to 2014-01-01
df = df[df['date'] >= '2015-01-01']

# Display the first 10 rows to verify
df.head(10)
```

I also want to reset the index back to zero to prevent any issues with merging down the line. 

```python
df.reset_index(drop=True, inplace=True)
df.head(10)
```

Now that I have the appropriate gold data in, I need to load in the bitcoin data. 

```python
#this is the API link for the Bitcoin data that I am using.
url = "https://eco-cors-proxy.netlify.app/proxy?url=https%3A%2F%2Fapi.stlouisfed.org%2Ffred%2Fseries%2Fobservations%3Fseries_id%3DCBBTCUSD%26api_key%3D22ee7a76e736e32f54f5df0a7171538d%26file_type%3Djson"

#this get the data out of the API url 
response = requests.get(url)
data = response.json()

#In the api, the data I want is under the heading of "observations" so i want to extract that and convert it to a dataframe 
df_observations = pd.DataFrame(data['observations'])

# Keep only the 'date' and 'value' columns
dfBTC = df_observations[['date', 'value']]
dfBTC.head(10)
```
