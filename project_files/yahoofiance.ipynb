```python
import yfinance as yf
```

First I will import the data on VIX index 

```python
#this defines the asset you want to look at then fetches the data and displays it in a dataframe 
ticker_symbol = '^VIX'

vix_data = yf.download(ticker_symbol, start='2016-01-01', end='2024-12-03')

vix_data.head(10)
```

When I began trying to filter some of the data, I was running into a lot of issued that stemmed from the formatting of the heading columns. so this code sorts the headings out and gets rid of the double stack headings. 

```python
print(vix_data.columns)
```

```python
vix_data.columns = vix_data.columns.get_level_values(0)
vix_data.reset_index(inplace=True)
```

```python

vix_data.columns = ['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']

vix_data = vix_data[['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']]
vix_data.head(10)
```

I also don't need all of this data, the only data I need is the data column and the Close column. 

```python
#this code keeps only those columns. 
vix_close = vix_data[['Date', 'Close']]
vix_close.head(10)
```

When trying to filter this data I was running into issues of NaN. The following code changes the format of the data I am using into to float form, so they are now defined as a number instead of a string, meaning it can be more easily filtered and altered. 

```python
vix_close['Close'] = vix_close['Close'].astype(float)
vix_close.head(10)
```

Now I need to begin filtering it to the data that I require for this chart. I want to find all the datapoints that are >25. I only need the values where it first hits 25 however, so any days where it is consecutively over 25 I want to remove. 

```python

# Filter to keep only the first instance where Close >= 30 and skip consecutive occurrences
filtered_vix_close = vix_close[(vix_close['Close'] >= 25) & 
                               (vix_close['Close'].shift(fill_value=0) < 25)]

filtered_vix_close.head(30)
```

Now we have the data we need for the VIX indicator I can load in the data for Bitcoin. 

```python
#this defines the asset you want to look at then fetches the data and displays it in a dataframe 
ticker2 = 'BTC-USD'

btc_data = yf.download(ticker2, start='2016-01-01', end='2024-12-03')

btc_data.head(10)
```

first I want to alter the column names as this caused lots of issues when managing the VIX data earlier. 

```python
# Flatten the headings so that its not across multiple levels
btc_data.columns = btc_data.columns.get_level_values(0)
print(btc_data.columns)

btc_data.reset_index(inplace=True)
```

```python

btc_data.columns = ['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']

btc_data = btc_data[['Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']]

btc_data.head(10)

```

I want to remove the unccessary columns. 

```python
btc_close = btc_data[['Date', 'Close']]
btc_close.head(10)
```

```python
filtered_vix_close.head(10)
```

Now I want to begin creating the merged dataframes to calcualte the percentage price changes for bitcoin in 1-day, 7-day, and 30-day periods

```python
#this merges the b


merged_1 = pd.merge(btc_close, filtered_vix_close, on='Date', how='inner')

merged_1 = merged_1.rename(columns={
    'Close_x': 'BTC_Close',
    'Close_y': 'VIX_Close'
})

merged_1.head(20)
```

```python
# Add a new column for BTC_Close value one day before the 'Date'
btc_close['Next_Day'] = btc_close['Date'] + pd.Timedelta(days=-1)

# Rename the 'Close' column for clarity
btc_close_renamed = btc_close.rename(columns={'Close': 'Next_Day_BTC_Close'})

# Create a new DataFrame with the merged data
merged_1day = pd.merge(
    merged_1,
    btc_close_renamed[['Next_Day', 'Next_Day_BTC_Close']],
    left_on='Date',
    right_on='Next_Day',
    how='left'
)

# Drop the auxiliary 'Next_Day' column from the new DataFrame to keep it clean
merged_1day = merged_1day.drop(columns=['Next_Day'])

# Display the first 20 rows of the new DataFrame
merged_1day.head(20)
```
