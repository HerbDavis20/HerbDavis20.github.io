```python
import pandas as pd
import requests
```

```python

df = pd.read_csv('BTCfiltere_data_intereactivechart.csv')
df
```

```python

df['BTC_Date'] = pd.to_datetime(df['BTC_Date'])

#group by vix date 
change30day = df.groupby('VIX_Date').apply(
    lambda x: ((x.iloc[-1]['BTC_Close'] - x.iloc[0]['BTC_Close']) / x.iloc[0]['BTC_Close']) * 100
).reset_index(name='change30day')

change30day.head(10)


change30day.to_csv('30day_percentagechange_fromspikechart.csv', index=False) 
change30day.head(10)
```
